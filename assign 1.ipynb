```python
1.
    
#Sort the list and find the max and min age

ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
ages.sort()
print("1. The sorted list of ages:", ages)
print("\tThe minimum age is:", min(ages))
print("\tThe maximum age is:", max(ages))

#Add the minimum age and the maximum age again to the list

ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
ages.append(19)
ages.append(26)
print("2. The final list is:", ages)

#Find the median age

import statistics as st
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24, 19, 26]
median_age = st.median(ages)
print("3. The median age of the list is:", median_age)

#Find the average age

from statistics import mean
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24, 19, 26]

mean_age= mean(ages)
print("4. The average age of the list is:", mean_age)

#Find the range of the ages (maximum minus min)

ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24, 19, 26]

maximum_age = maximum(ages)
minimum_age = minimum(ages)
range = maximum_age - minimum_age
print("5. The range of the ages is:", range)
```

```python
6.

#“I am a teacher and I love to inspire and teach people”

#How many unique words have been used in the sentence? Use the split methods and set to get the unique word
string = "I am a teacher and I love to inspire and teach people"
new_words = string.split(" ")
n = 0
for word in new_words:
    word_count = new_words.count(word)
    if word_count == 1:
        n += 1
print("The number of unique words:", n)
```
